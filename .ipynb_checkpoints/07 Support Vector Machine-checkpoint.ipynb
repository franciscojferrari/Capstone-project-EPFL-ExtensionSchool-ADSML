```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import glob
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

## Load the Data and split in Train and Test Datasets

```python
df_final = pd.read_excel('Final Dataset.xlsx', sheet_name='data')

features = df_final[['n_pores', 'median_log_pores', 'n_agg', 'median_log_agg', 'r', 'g', 'b']]
img_name = df_final.iloc[:,1:2]

spec = df_final.iloc[:,0:1]
spec = spec.replace({'in-spec':1, 'out-of-spec':0})


f_tr, f_te, c_tr, c_te = train_test_split(
    features, spec, test_size=0.3, stratify=spec, random_state=1)
```

# SVM with Linear Kernel

### Model Tuning

```python
from sklearn.model_selection import StratifiedKFold

# Create k-fold object
kfold = StratifiedKFold(n_splits=5)
```

```python
from sklearn.model_selection import ParameterGrid
from sklearn.model_selection import cross_validate

C = np.logspace(-2, 6, 9)
#WE LIMIT THE NUMBER FOR C VALUES DUE TO THE PERFORMANCE ISSUES.

# Save accuracy on test set
test_scores_mean = []
test_scores_std = []

pipeLSVC= Pipeline([
#     ('scaler', StandardScaler()),
    ('scaler', MinMaxScaler()),
#     ('LSVC', LinearSVC()
    ('LSVC', SVC(kernel='linear',probability=True)
    )])

# Define a grid of values
grid = ParameterGrid({
    'LSVC__C': C,
})

for params_dict in grid:
    # Set parameters
    pipeLSVC.set_params(**params_dict)   
            
    rf_scores = cross_validate(pipeLSVC, f_tr, c_tr, cv=kfold)
    
    test_scores_mean.append(np.mean(rf_scores['test_score']))
    test_scores_std.append(np.std(rf_scores['test_score']))

    
linear_svm = pd.DataFrame.from_dict({"C":C, "Mean":test_scores_mean, "Std Dev": test_scores_std})
linear_svm.sort_values(by='Mean', ascending=False)
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,8))
plt.semilogx(linear_svm['C'], linear_svm['Mean'])
plt.xlabel('Mean')
plt.ylabel('C')
plt.show()
```

```python
linear_svm_top = linear_svm.sort_values(by='Mean', ascending=False).head(1)
linear_svm_top
```

# SVM with Non-Linear Kernel

```python
from sklearn.model_selection import ParameterGrid
from sklearn.model_selection import GridSearchCV

# Save accuracy on test set
test_svc_scores_mean = []
test_svc_scores_std = []
params_svc = []

#Define Values of C and Gamma
gamma = np.logspace(-7, 3, 11)
C = np.logspace(-2, 8, 11)



pipeSVC= Pipeline([
#     ('scaler', StandardScaler()),
    ('scaler', MinMaxScaler()),
    ('SVC', SVC()
    )])

grid_svc = GridSearchCV(pipeSVC, [{
    'SVC__gamma': gamma,
    'SVC__C': C
}], cv=kfold)
grid_svc.fit(f_tr, c_tr)

# Collect results in a DataFrame
nonlinear_svm = pd.DataFrame.from_items([
    ('C', grid_svc.cv_results_['param_SVC__C']),
    ('gamma', grid_svc.cv_results_['param_SVC__gamma']),
    ('mean', grid_svc.cv_results_['mean_test_score']),
    ('std', grid_svc.cv_results_['std_test_score'])
])
nonlinear_svm.sort_values(by='mean', ascending=False)

```

```python
c_gamma_results = nonlinear_svm.iloc[:,0:3]
c_gamma_results = c_gamma_results.pivot("gamma", "C", "mean")

c_gamma_std = nonlinear_svm.iloc[:,[0,1,3]]
c_gamma_std = c_gamma_std.pivot("gamma", "C", "std")

f, ax = plt.subplots(1,2,figsize=(20,7))
sns.heatmap(c_gamma_results, annot=True, ax=ax[0])
sns.heatmap(c_gamma_std, annot=True, ax=ax[1])
```

```python
nonlinear_svm_top = nonlinear_svm.sort_values(by='mean', ascending=False).head(1)
```

```python
print("Linear SVM - top accuracy across folds: {:.3f} (std: {:.3f}) with C:{:.3f}".format(
    linear_svm_top.iloc[0]['Mean'] * 100,
    linear_svm_top.iloc[0]['Std Dev'],
    linear_svm_top.iloc[0]['C']
))
print("RBF SVM - top accuracy across folds: {:.3f} (std: {:.3f} with C: {:.3f} and gamma: {:.3f}".format(
    nonlinear_svm_top.iloc[0]['mean'] * 100,
    nonlinear_svm_top.iloc[0]['std'],
    nonlinear_svm_top.iloc[0]['C'],
    nonlinear_svm_top.iloc[0]['gamma']
))
```

### As we are using cross validation we can appreciate that for high values of gamma we are having a low mean scores since the model is probably overfeating for each data sample being used. <br> Larger values of C have a larger standar derivation signs that model could be overfiting

# Evaluate accuracy

```python
# We run the tuned Linear SVM 
pipeLSVC= Pipeline([
    ('scaler', MinMaxScaler()),
#     ('LSVC', LinearSVC(random_state=0, C=linear_svm_top.iloc[0]['C'])
    ('LSVC', SVC(kernel='linear',probability=True, C=linear_svm_top.iloc[0]['C'])
    )])

#Fit train data
pipeLSVC.fit(f_tr, c_tr)

# Evaluate on test data
accuracy = pipeLSVC.score(f_te, c_te) * 100
print('Linear SVM accuracy (test set): {:.3f} %'.format(accuracy))
```

```python
# We run the tuned Non-Linear SVM 
pipeSVC = Pipeline([
    ('scaler', MinMaxScaler()),
    ('SVC', SVC(C=nonlinear_svm_top.iloc[0]['C'],gamma=nonlinear_svm_top.iloc[0]['gamma'])
    )])

#Fit train data
pipeSVC.fit(f_tr, c_tr)

# Evaluate on test data
accuracy = pipeSVC.score(f_te, c_te) * 100
print('RBF SVM accuracy (test set): {:.3f} %'.format(accuracy))
```
