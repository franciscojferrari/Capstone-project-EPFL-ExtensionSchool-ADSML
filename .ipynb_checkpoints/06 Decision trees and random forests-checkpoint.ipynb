```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ParameterGrid
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
```

## Load the Data and split in Train and Test Datasets

```python
df_final = pd.read_excel('Final Dataset.xlsx', sheet_name='data')

features = df_final[['n_pores', 'median_log_pores', 'n_agg', 'median_log_agg', 'r', 'g', 'b']]
img_name = df_final.iloc[:,1:2]

spec = df_final.iloc[:,0:1]
spec = spec.replace({'in-spec':1, 'out-of-spec':0})


f_tr, f_te, c_tr, c_te = train_test_split(
    features, spec, test_size=0.3, stratify=spec, random_state=1)

# # We split the train set in train and validate
f_tr, f_va, c_tr, c_va = train_test_split(
    f_tr, c_tr, test_size=0.1,stratify=c_tr, random_state=1)
```

# Decision Trees

```python
max_depth = np.arange(1, 30) 

#to save dataset accuracy
test_scores_tr = []
test_scores_va = []


for i in max_depth:
    # Set parameters
    pipe = Pipeline([
        ('scaler', MinMaxScaler()),
        ('Dt', DecisionTreeClassifier( max_depth = max_depth[ (i - 1) ], random_state=0) )
    ])
    
    # Fit classifier
    pipe.fit(f_tr, c_tr)
   
    # Save result
    test_scores_tr.append(pipe.score(f_tr, c_tr))
    test_scores_va.append(pipe.score(f_va, c_va))
```

```python
# Now we plot and compare the train and validation accuracy

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure()
ax = plt.axes()

plt.plot(max_depth, test_scores_va, label='Validation Set')
plt.plot(max_depth, test_scores_tr, label='Train Set')
ax.legend()
plt.ylabel("Accuracy")
plt.xlabel("Depth")
plt.title("Accuracy by Depth")

plt.show()
```

```python
test_scores_va_pd = pd.DataFrame(test_scores_va, columns=['Validation'] )

print("Top accuracy on validation set: {:.2f}% with depth {}".format(
    test_scores_va_pd['Validation'].max() * 100, 
    max_depth[test_scores_va_pd['Validation'].idxmax()])
     )
tree_depth = max_depth[test_scores_va_pd['Validation'].idxmax()]
```

#### We can clearly apreciate the overfitting on the test set as the depth increases <br> We can also appreciate that the accuracy in the validation set keeps flat this is due probably to the limited size of the dataset


## Let's investigate what are the features that are driving the classification

```python
from sklearn.tree import export_graphviz
import graphviz


dt2 = DecisionTreeClassifier(max_depth=2, random_state=0)

# Fit decision tree
dt2.fit(f_tr, c_tr)
# Get score
dt2.score(f_tr, c_tr)

# Export decision tree
dot_data = export_graphviz(
    dt2, out_file=None,
    class_names = ['out-of-spec', 'in-spec'],
    feature_names=features.columns,
    filled=True, rounded=True, proportion=True   
)
# Display decision tree
graphviz.Source(dot_data)
    
```

```python
dt3 = DecisionTreeClassifier(max_depth=3, random_state=0)

# Fit decision tree
dt3.fit(f_tr, c_tr)
# Get score
dt3.score(f_tr, c_tr)

# Export decision tree
dot_data = export_graphviz(
    dt3, out_file=None,
    class_names = ['out-of-spec', 'in-spec'],
    feature_names=features.columns,
    filled=True, rounded=True, proportion=True   
)
# Display decision tree
graphviz.Source(dot_data)
```

### We can see that the main features are the number of aggregation and pores and the color channels.

# Random Forest

```python
# Save accuracy
accuracy_tr_forest = []
accuracy_va_forest = []
features_importance = pd.DataFrame(columns=features.columns)
inter_tree_variability = pd.DataFrame(columns=features.columns)

t_s = []
m_d = []

tree_size = np.arange(10, 200, 10)
max_depths = np.arange(1, 10, 1)
max_depths = np.append(max_depths, [None])

for j in max_depths:
    for i in tree_size:
        # Set parameters
        pipe = Pipeline([
            ('scaler', MinMaxScaler()),
            ('rf', RandomForestClassifier(n_estimators = i, max_depth = j))
        ])
        # Fit classifier
        pipe.fit(f_tr, c_tr)

        importance = pipe.steps[1][1].feature_importances_
        importance = importance.reshape(-1, len(importance))
        
        std = np.std([tree.feature_importances_ for tree in pipe.steps[1][1].estimators_],
             axis=0)
        std = std.reshape(-1, len(std))

        features_importance = features_importance.append(pd.DataFrame(importance, columns=features.columns), ignore_index=True)
        inter_tree_variability = inter_tree_variability.append(pd.DataFrame(std, columns=features.columns), ignore_index=True)

        # Save result
        t_s.append(i)
        m_d.append(j)
        accuracy_tr_forest.append(pipe.score(f_tr, c_tr))
        accuracy_va_forest.append(pipe.score(f_va, c_va))
```

```python
#We get the best values
accuracy_tr_forest_pd = pd.DataFrame(accuracy_tr_forest, columns=['Train'] )
accuracy_va_forest_pd = pd.DataFrame(accuracy_va_forest, columns=['Validation'] )
tree_size_pd = pd.DataFrame(t_s, columns=['N estimators'] )
max_depths_pd = pd.DataFrame(m_d, columns=['Max Depths'] )
scores = pd.concat([max_depths_pd, tree_size_pd, accuracy_tr_forest_pd, accuracy_va_forest_pd], axis=1)

results = scores.sort_values(by='Validation', ascending=False).head(1)
m_depth = results['Max Depths'].values
m_depth = m_depth[0]
n_stimators = results['N estimators'].values
n_stimators = n_stimators[0]

scores.sort_values(by='Validation', ascending=False).head(10)
```

```python
rf_results_tr = scores.iloc[:,0:3]
rf_results_tr = rf_results_tr.pivot("Max Depths", "N estimators", "Train")

rf_results_va = scores.iloc[:,[0,1,3]]
rf_results_va = rf_results_va.pivot("Max Depths", "N estimators", "Validation")

f, ax = plt.subplots(2,1,figsize=(20,15))

ax[0].set_title('Training set accuracy',fontsize=20 )
sns.heatmap(rf_results_tr, annot=True, ax=ax[0])

ax[1].set_title('Validation set accuracy',fontsize=20 )
sns.heatmap(rf_results_va, annot=True, ax=ax[1])
```

#### We can see how there is an overfiting on the training set with max depth set to None or high numbers. <br> For the validation data set setting the max_depth to None (max) doesn't have a positive impact

## Let's plot the most important features of the top model

```python
# Plot the feature importances of the forest
# plt.figure()
# plt.title("Feature importances")

most_important = features_importance.iloc[results.index[0]]

ax = most_important.plot(kind='bar', figsize=(15, 7))
ax.set_ylabel("importance", fontsize=12)
plt.show()
```

### As with the single decistion tree the most important features for the random forest are the number of pores and aggregations and the red channel

```python
pipe = Pipeline([
    ('scaler', MinMaxScaler()),
    ('rf', RandomForestClassifier(n_estimators = n_stimators, max_depth = m_depth))
])

# Fit to train data
pipe.fit(f_tr, c_tr)

print('Random forest accuracy (test set): {:.2f}%'.format(pipe.score(f_te, c_te) * 100))
```
